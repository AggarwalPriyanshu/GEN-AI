```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
model_name='gpt2'

tokenizer=GPT2Tokenizer.from_pretrained(model_name)
model=GPT2LMHeadModel.from_pretrained(model_name)
```

```python
def generate_POEM(prompt,max_length=100):

  #encoding the prompt using our tokenizer
  input_ids=tokenizer.encode(prompt,return_tensors='pt')

  #generating the poetry using our model
  Output=model.generate(input_ids,
                        max_length=max_length,
                        num_return_sequences=1,
                        no_repeat_ngram_size =2,
                        top_k=50,
                        top_p=0.95,
                        temperature=0.7)

  generated_text = tokenizer.decode(Output[0],skip_special_tokens=True)
  return generated_text
```

```python
prompt = "Twinkle Twinkle Litter Star "

poem  = generate_POEM(prompt)

print("---------------------POEM-----------------------")
print(poem)
```

```python
from music21 import *
import random


def generate_melody(length=8 , scale=scale.MajorScale('C')):
  melody = stream.Stream()

  for _ in range (length):

    pitch = random.choice(scale.getPitches("C4","C5"))

    #QUARTER , HALF , DOTTED HALF , WHOLE NOTE

    duration = random.choice([0.5,1,1.5,2])

    new_note=note.Note(pitch)
    new_note.duration.quarterLength=duration

    melody.append(new_note)

  return melody
```

```python
melody = generate_melody()

melody.show('midi')
```

PART-3 COMBINING POETRY AND MUSIC


WE WILL TRANSLATE POETRY TO MUSIC

```python
def poem_to_melody(poem,scale = scale.MajorScale("A")):

  words = poem.split()
  melody = stream.Stream()

  for work in words[:8]: #limit to 8 notes for simplicity

    pitch = scale.pitchFromDegree((len(words)&7) + 1)

    duration = len(words) * 0.25 #quarter node for each letter

    new_note=note.Note(pitch)

    new_note.duration.quarterLength = duration

    melody.append(new_note)

    return melody
```

```python
poem = generate_POEM("THE WIND WHISPERS,")

print("---------------------------------POEM------------------------------------")
print(poem)


print("---------------------------------------------------------------------------")

melody=poem_to_melody(poem)
melody.show('midi')
```
